```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("imdb_top_1000.csv")
data.head()
```

```python
data_shape=data.shape
data_size=data.size
print(data_shape)
print(data_size)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
cols = ['Poster_Link','Released_Year','Certificate','Runtime','Overview']
```

```python
data.drop(columns=cols,inplace=True)
data
```

```python
data['Genre'].value_counts()
post_popular_genre =data['Genre'].value_counts().head(1)
print(post_popular_genre)
```

```python
Genre_df = data.Genre.value_counts()
Genre_df
```

```python
import matplotlib.pyplot as plt

movies = Genre_df[0:10].index
count = list(Genre_df.iloc[:10])

plt.barh(movies,count)
plt.ylabel("Most Popular Genres")
plt.xlabel("Popularity")
plt.title("Top 10 Catefories")
plt.show()
```

```python
Director = data.Director.value_counts()
Director
```

```python
directors = Director[0:10].index
counter = list(Director.iloc[:10])
plt.barh(directors,counter)
plt.ylabel("Best Directors")
plt.xlabel("high rated movies")
plt.title("Top 10 Directors")
plt.show()
```
